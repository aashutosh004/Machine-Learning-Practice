```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
X,y = load_iris(return_X_y = True)
```

```python
X
```

```python
y
```

```python
X_train ,X_test , y_train , y_test = train_test_split(X,y,test_size = 0.3,random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train,y_train)
y_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
import seaborn as sns
df=sns.load_dataset('tips')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df.head()
```

```python
## time as target

X = df[['total_bill','tip', 'sex', 'smoker', 'day', 'size']]
y = df['time']
```

```python
X
```

```python
y
```

```python
## Train Test Split

from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.25 , random_state = 10)
```

```python
X_train.head()
```

```python
# 1. Label encoding

from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()

X_train['sex'] = le1.fit_transform(X_train['sex'])
X_train['smoker'] = le2.fit_transform(X_train['smoker'])
```

```python
X_test['sex'] = le1.transform(X_test['sex'])
X_test['smoker'] = le2.transform(X_test['smoker'])
```

```python
X_train
```

```python
## One hot encoding

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers = [('onehot',OneHotEncoder(drop = 'first'),[4])],remainder = 'passthrough')
import sys
import numpy as np
np.set_printoptions(threshold = sys.maxsize)
X_train = ct.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = ct.transform(X_test)
X_test
```

```python
## Label encoding of target y
le3 = LabelEncoder()
y_train = le3.fit_transform(y_train)
```

```python
y_test = le3.transform(y_test)
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train,y_train)
y_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
