```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
from matplotlib import pyplot
```

```python
X , y = make_classification(n_samples = 1000 , n_classes = 2 , random_state = 1)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y ,test_size = 0.25 , random_state = 0)
```

```python
# generate a no skill predcition (majority class)

dummy_model_prob = [0 for _ in range(len(y_test))]
dummy_model_prob
# we have created a dummy model with default as 0 output
```

```python
# logistic model

model = LogisticRegression()
model.fit(X_train , y_train)
```

```python
model_prob = model.predict_proba(X_test)
```

```python
model_prob
```

```python
model_prob = model_prob[:,1] # focus on the positive outcome
```

```python
model_prob
```

```python
# calculate the roc_auc_curve

dummy_model_auc = roc_auc_score(y_test , dummy_model_prob)
model_auc = roc_auc_score(y_test , model_prob)

print(dummy_model_auc)
print(model_auc)
```

```python
# calculate ROC Curves

dummy_fpr , dummy_tpr , _ = roc_curve(y_test , dummy_model_prob)
model_fpr , model_tpr , thresholds = roc_curve(y_test , model_prob)
```

```python
# plotting roc curve

pyplot.plot(dummy_fpr , dummy_tpr , linestyle='--',label = 'Dummy Model')
pyplot.plot(model_fpr , model_tpr , marker = '.',label = 'Logistic')

pyplot.xlabel('False Positive Rate')
pyplot.xlabel('True Positive Rate')

pyplot.legend()

pyplot.show()
```

```python
import numpy as np

# Create figure and subplot
fig = pyplot.figure(figsize=(20, 50))  # Adjust size if needed
ax = fig.add_subplot(111)

# Plot the lines for both models
ax.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy Model')
ax.plot(model_fpr, model_tpr, marker='.', label='Logistic')

# Annotate thresholds
for xyz in zip(model_fpr, model_tpr, thresholds):
    ax.annotate("%s" % np.round(xyz[2], 2), xy=(xyz[0], xyz[1]))

# Set labels and legend
ax.set_xlabel("FPR")
ax.set_ylabel("TPR")
ax.legend()


# Show the plot
pyplot.show()
```
